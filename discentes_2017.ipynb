Análise de discentes de Gestão hospitalar 2017.2, para saber o atual status dos alunos atualmente. 

```python
import pandas as pd
discentes_2017 = pd.read_csv('discentes-2017.csv', sep=';')
discentes_2017.head()
```

```python
gestaoHospitalar_2 = discentes_2017[(discentes_2017.id_curso == 2000159) & (discentes_2017.periodo_ingresso == 2)]
```

```python
gestaoHospitalar_2.head()
```

```python
gestaoHospitalar_2[['nome_discente','status']]
```

```python
statusGH17_2 = gestaoHospitalar_2['status'].value_counts()
```

```python
statusGH17_2
```

```python
type(statusGH17_2)
```

```python
# import numpy as np
import matplotlib.pyplot as plt
```

```python
dfGH = pd.DataFrame(statusGH17_2)

dfGH.head()
```

```python
plt.figure(figsize=(7,7))

plt.pie(x = dfGH['status'], labels=dfGH.index, shadow=False, startangle=90, autopct= '%1.1f%%')
plt.title('Status dos discentes de Gestão Hospitalar 2017.2 atualmente(Março/2022)')

plt.show()
```
